```python
from IPython.display import clear_output

def display_board(board):
    
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('_____')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('_____')
    print(board[1] + '|' + board[2] + '|' + board[3])

def player_input():
    choice = input('Choose X or O: ')
    while choice !='X' and choice!= 'O':
        choice = input('Please enter correct choice')
        
    
    if choice == 'X':
        print('Player 1 has to play now')
        player1 = 'X'
        player2 = 'O'
    else:
        print('Player 2 has to play now')
        player2 = 'X'
        player1 = 'O'
    return (player1,player2)

def place_marker(board, marker, position):
    board[position] = marker
    
def win_check(b, mark):
    return b[7]==b[8]==b[9]== mark or b[4]==b[5]==b[6]== mark or b[1]==b[2]==b[3]== mark or b[1]==b[4]==b[7]== mark or b[2]==b[5]==b[8]== mark or b[3]==b[6]==b[9]== mark or b[1]==b[5]==b[9]== mark or b[7]==b[5]==b[3]== mark

def space_check(board, position):
    return board[position] == ' '

def full_board_check(board):
    ran = range(1,len(board))
    for ind in ran:
        if board[ind] == ' ':
            return False
    else:
        return True

def player_choice(board):
    num = int(input('Enter a number from 1-9: '))
    booll = space_check(board,num)
    
    while booll == False:
        num = int(input('Already occupied...Enter a number from 1-9: '))
        booll = space_check(board,num)
    
    return num

def replay():
    play_again = input('Wanna play again? Type in yes or no: ')
    while play_again.lower()!= 'yes' and play_again.lower()!='no':
        play_again = input('Please enter the correct reply. Type in yes or no: ')
    if play_again.lower() == 'yes':
        return True
    else:
        return False

print('Welcome to Tic Tac Toe!')

board_list = [' ']*10

#p1_marker, p2_marker = player_input()
game = True

while game == True:
    p1_marker, p2_marker = player_input()
    while full_board_check(board_list) == False and win_check(board_list,p1_marker) == False and win_check(board_list,p2_marker) == False:
        
        if p1_marker == 'X':
            print('Player 1, aadra dei!!')
            entered_number = player_choice(board_list)
            place_marker(board_list, p1_marker, entered_number)
            clear_output()
            display_board(board_list)
            if win_check(board_list,p1_marker) == True or full_board_check(board_list) == True:
                continue
            print('Player 2, nee adra!')
            entered_number = player_choice(board_list)
            place_marker(board_list, p2_marker, entered_number)
            clear_output()
            display_board(board_list)
            if win_check(board_list,p2_marker) == True or full_board_check(board_list) == True:
                continue
            
        else:
            print('Player 2, aadra dei!!')
            entered_number = player_choice(board_list)
            place_marker(board_list, p2_marker, entered_number)
            clear_output()
            display_board(board_list)  
            if win_check(board_list,p2_marker) == True or full_board_check(board_list) == True:
                continue
            print('Player 1, nee adra!')
            entered_number = player_choice(board_list)
            place_marker(board_list, p1_marker, entered_number)
            clear_output()
            display_board(board_list)
            if win_check(board_list,p1_marker) == True or full_board_check(board_list) == True:
                continue
            

    if win_check(board_list,p1_marker) == True:
        print('Player 1 has won the game!!!')
    elif win_check(board_list,p2_marker) == True:
        print('Player 2 has won the game!!!')
    elif full_board_check(board_list) == True:
        print('The game is tied da!!!')

    out = replay()
    if out == True:
        board_list = [' ']*10
    else:
        print('Thanks da dei for playing!!')
        break


    
```
